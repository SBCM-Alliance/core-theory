```python
# ==============================================================================
# ðŸ›ï¸ SBCM Collider: Global Edition (Interactive Ver.)
# ------------------------------------------------------------------------------
# Author: SBCM Alliance
# Target: Google Colab Native
# ==============================================================================

# 1. å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã¿ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ï¼ˆkaleidoã¯é™¤å¤–ï¼‰
!pip install pandas plotly

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# ==============================================================================
# 2. SBCM Constants & Real Data Loader
# ==============================================================================
B_STD = 72176       # Standard Block Population
F_BORDER = 0.70     # Fiscal Autonomy Threshold
D_BORDER = 1.20     # Distortion Limit

def load_global_data():
    # Real data from Japanese municipalities (Pop â‰ˆ 70k)
    cities = [
        # --- Theme A: Colony / Bedtown ---
        {"name": "Moriya (Ibaraki)", "pop": 69000, "fiscal": 1.02, "burden": 10.5, "civil_eng": 12.0, "type": "Colony"},
        {"name": "Nagakute (Aichi)", "pop": 60000, "fiscal": 1.03, "burden": 0.0, "civil_eng": 15.0, "type": "Colony"},
        {"name": "Natori (Miyagi)",  "pop": 79000, "fiscal": 0.78, "burden": 20.0, "civil_eng": 14.0, "type": "Colony"},

        # --- Theme B: Heat Death / Decay ---
        {"name": "Odate (Akita)",    "pop": 66000, "fiscal": 0.42, "burden": 18.0, "civil_eng": 22.0, "type": "HeatDeath"},
        {"name": "Tsuruga (Fukui)",  "pop": 63000, "fiscal": 0.75, "burden": 80.0, "civil_eng": 25.0, "type": "HeatDeath"},
        {"name": "Miyakojima (Okinawa)", "pop": 55000, "fiscal": 0.45, "burden": 60.0, "civil_eng": 20.0, "type": "HeatDeath"},
        {"name": "Maizuru (Kyoto)",  "pop": 77000, "fiscal": 0.50, "burden": 90.0, "civil_eng": 18.0, "type": "HeatDeath"},

        # --- Theme C: Junction / Pump ---
        {"name": "Tosu (Saga)",      "pop": 74000, "fiscal": 0.90, "burden": 30.0, "civil_eng": 13.0, "type": "Junction"},
        {"name": "Kameyama (Mie)",   "pop": 50000, "fiscal": 0.95, "burden": 40.0, "civil_eng": 16.0, "type": "Junction"},

        # --- Others (Comparison) ---
        {"name": "Tenri (Nara)",     "pop": 63000, "fiscal": 0.55, "burden": 120.0, "civil_eng": 15.0, "type": "Zombie"},
        {"name": "Chichibu (Saitama)", "pop": 58000, "fiscal": 0.50, "burden": 30.0, "civil_eng": 18.0, "type": "Zombie"},
        {"name": "Minamisoma (Fukushima)", "pop": 59000, "fiscal": 0.48, "burden": 10.0, "civil_eng": 30.0, "type": "HeatDeath"},
    ]

    df = pd.DataFrame(cities)
    # Calculate SBCM Distortion Index
    df['SBCM_Distortion'] = (df['civil_eng'] / 10.0) / (df['fiscal'] * 0.8)
    return df

# ==============================================================================
# 3. Diagnostic Engine
# ==============================================================================
def diagnose(row):
    f = row['fiscal']
    d = row['SBCM_Distortion']

    if f >= F_BORDER:
        if d < D_BORDER:
            return "Q1: Autonomy (Ideal)"
        else:
            return "Q2: Colony (Extraction)"
    else:
        if d < D_BORDER:
            return "Q3: Stagnation (Zombie)"
        else:
            return "Q4: Heat Death (Collapse)"

# ==============================================================================
# 4. Visualization (Interactive)
# ==============================================================================
def plot_sbcm_interactive(df):
    df['Diagnosis'] = df.apply(diagnose, axis=1)
    # ãƒãƒ–ãƒ«ã‚µã‚¤ã‚ºèª¿æ•´
    df['BubbleSize'] = df['burden'].apply(lambda x: max(x, 15))

    fig = px.scatter(
        df,
        x="fiscal",
        y="SBCM_Distortion",
        size="BubbleSize",
        color="Diagnosis",
        hover_name="name",
        text="name",
        title="<b>SBCM Matrix: Thermodynamic Analysis of 70k Cities</b>",
        labels={
            "fiscal": "Fiscal Strength Index (Input Energy) â†’",
            "SBCM_Distortion": "SBCM Distortion Index (Structural Strain) â†‘",
            "burden": "Future Burden Ratio",
            "Diagnosis": "SBCM Status"
        },
        color_discrete_map={
            "Q1: Autonomy (Ideal)":      "#2ecc71", # Green
            "Q2: Colony (Extraction)":   "#f1c40f", # Yellow
            "Q3: Stagnation (Zombie)":   "#95a5a6", # Gray
            "Q4: Heat Death (Collapse)": "#e74c3c"  # Red
        },
        range_x=[0.2, 1.2],
        range_y=[0.5, 4.0],
        width=1000,
        height=700,
        template="plotly_white"
    )

    # Threshold Lines
    fig.add_vline(x=F_BORDER, line_dash="dash", line_color="gray", annotation_text="Autonomy Line")
    fig.add_hline(y=D_BORDER, line_dash="dash", line_color="gray", annotation_text="Distortion Limit")

    # Annotations
    fig.add_annotation(x=1.1, y=0.8, text="<b>Q1: Ideal State</b>", showarrow=False, font=dict(color="green", size=14))
    fig.add_annotation(x=0.3, y=3.5, text="<b>Q4: HEAT DEATH</b>", showarrow=False, font=dict(color="red", size=14))

    fig.update_traces(textposition='top center', textfont_size=11)
    fig.update_layout(legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01))

    # Colabä¸Šã§ã‚¤ãƒ³ã‚¿ãƒ©ã‚¯ãƒ†ã‚£ãƒ–è¡¨ç¤º
    fig.show()

# ==============================================================================
# Execution
# ==============================================================================
df_real = load_global_data()
plot_sbcm_interactive(df_real)

print("\n--- Diagnostic Report (Top 5 High Distortion) ---")
print(df_real.sort_values("SBCM_Distortion", ascending=False).head(5)[['name', 'fiscal', 'SBCM_Distortion']])
```
