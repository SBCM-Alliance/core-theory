```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter Settings
A = np.linspace(1, 100, 1000)
sigma = 1.5   # Productivity exponent (Singularity logic)
gamma = 2.2   # Complexity penalty exponent (Governance Engineering logic)

# Models
singularity_path = A**sigma           # Infinite growth model
heat_death_path = A**sigma - 0.05*A**gamma  # GE model with complexity penalty

# Plotting
plt.figure(figsize=(10, 8))
plt.plot(A, singularity_path, label='Virtual Singularity (Macro-Models)',
         color='blue', linestyle='--', linewidth=2)
plt.plot(A, heat_death_path, label='Thermodynamic Reality (Governance Engineering)',
         color='red', linewidth=3)

# Add "The Thermal Wall" point
wall_idx = np.argmax(heat_death_path)
plt.annotate('THE THERMAL WALL', xy=(A[wall_idx], heat_death_path[wall_idx]),
             xytext=(A[wall_idx]-20, heat_death_path[wall_idx]+200),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=16, fontweight='bold', color='red')

plt.title('The Paradox of Intelligence: Singularity vs. Heat Death', fontsize=16, fontweight='bold')
plt.xlabel('Intelligence / Complexity (A)', fontsize=12)
plt.ylabel('Net Systemic Work (Output - Entropy Cost)', fontsize=12)
plt.ylim(0, 1500)
plt.legend()
plt.grid(True, alpha=0.3)

# Style as a "Cover Art"
plt.fill_between(A, singularity_path, heat_death_path, color='grey', alpha=0.1, label='Entropy Gap')

plt.show()
```
