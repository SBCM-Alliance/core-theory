```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set style for academic paper
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 10

# 1. Resource Mass Dynamics (FY2024 vs FY2025)
# Data from Page 1: Growth rate by organization type
mass_data = {
    'Org Type': ['Total', 'Prefectures', 'Designated Cities', 'Cities', 'Towns/Villages', 'Cooperatives'],
    'Increase_Rate': [3.5, 2.7, 7.6, 3.5, 2.0, 2.6]
}
df_mass = pd.DataFrame(mass_data)

# 2. Functional Mass Distribution (Job Type)
# Data from Page 3: Where the energy is directed
job_data = {
    'Job Type': ['General Clerical', 'Nursery Teachers', 'Skilled Laborers', 'Teachers/Lecturers', 'Nurses/Medics', 'Others'],
    'Count': [227104, 97468, 94810, 35888, 32106, 153661 + 7981 + 5320 + 12577 + 19721] # Aggregated others
}
df_job = pd.DataFrame(job_data).sort_values('Count', ascending=True)

# 3. Workload Intensity (Weekly Hours)
# Data from Page 5: Temporal density of the staff
hour_data = {
    'Hours Range': ['19.4h - 23.25h', '23.25h - 31h', '31h - 37.5h', '37.5h+'],
    'Percentage': [11.0, 49.7, 29.1, 10.2]
}
df_hour = pd.DataFrame(hour_data)

# Create Charts
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot 1: Mass Growth Rate (Structural Complexity)
colors1 = ['#e74c3c' if x > 5 else '#3498db' for x in df_mass['Increase_Rate']]
axes[0].bar(df_mass['Org Type'], df_mass['Increase_Rate'], color=colors1)
axes[0].set_title('Fig 4. Mass Growth Rate by Entity Type\n(FY2024 to FY2025)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Percentage Increase (%)')
axes[0].tick_params(axis='x', rotation=45)

# Plot 2: Functional Distribution (Energy Allocation)
axes[1].barh(df_job['Job Type'], df_job['Count'], color='#2ecc71')
axes[1].set_title('Fig 5. Functional Mass Distribution\n(Job Types in FY2025)', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Number of Staff (Mass)')

# Plot 3: Workload Intensity (Temporal Density)
axes[2].pie(df_hour['Percentage'], labels=df_hour['Hours Range'], autopct='%1.1f%%',
        startangle=140, colors=['#f1c40f', '#e67e22', '#d35400', '#c0392b'], explode=(0, 0.1, 0, 0))
axes[2].set_title('Fig 6. Workload Intensity\n(Weekly Working Hours Distribution)', fontsize=12, fontweight='bold')

plt.tight_layout()

# Save figures to files
fig.savefig('combined_charts.png') # Save all subplots in one file

# If you want to save each subplot separately, you would do:
fig1, ax1 = plt.subplots(figsize=(6, 6))
ax1.bar(df_mass['Org Type'], df_mass['Increase_Rate'], color=colors1)
ax1.set_title('Fig 4. Mass Growth Rate by Entity Type\n(FY2024 to FY2025)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Percentage Increase (%)')
ax1.tick_params(axis='x', rotation=45)
plt.tight_layout()
fig1.savefig('mass_growth_rate_fig4.png')

fig2, ax2 = plt.subplots(figsize=(6, 6))
ax2.barh(df_job['Job Type'], df_job['Count'], color='#2ecc71')
ax2.set_title('Fig 5. Functional Mass Distribution\n(Job Types in FY2025)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Number of Staff (Mass)')
plt.tight_layout()
fig2.savefig('functional_distribution_fig5.png')

fig3, ax3 = plt.subplots(figsize=(6, 6))
ax3.pie(df_hour['Percentage'], labels=df_hour['Hours Range'], autopct='%1.1f%%',
        startangle=140, colors=['#f1c40f', '#e67e22', '#d35400', '#c0392b'], explode=(0, 0.1, 0, 0))
ax3.set_title('Fig 6. Workload Intensity\n(Weekly Working Hours Distribution)', fontsize=12, fontweight='bold')
plt.tight_layout()
fig3.savefig('workload_intensity_fig6.png')

# Removed plt.show() as per user's request to save images instead of displaying

# Print Table for Paper Citation (AD converted)
print("--- Data Summary for Governance Engineering Paper (Gregorian Calendar) ---")
print(f"Total Temporary/Part-time Mass (FY2025): 768,752 (+3.5% from FY2024)")
print(f"Highest Complexity Growth: Designated Cities (+7.6%)")
print(f"Dominant Workload Mode: 23.25h - 31h/week (49.7% of total part-time mass)")
```
