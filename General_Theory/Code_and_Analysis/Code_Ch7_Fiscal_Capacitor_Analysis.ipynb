```python
import matplotlib.pyplot as plt
import pandas as pd

# Set academic style
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.size'] = 11

# Data from Chapter 7.2 and 7.3 (FY2024 Estimates)
entities = ['Tokyo\n(The Capacitor)', 'Osaka\n(The Debtor)', 'Aichi\n(The Maker)']
tax_dependency = [74.8, 47.7, 51.8]  # R_tax (%)
fund_rotation = [0.47, 0.15, 0.12]   # R_fund (Ratio)

# Create Charts
fig, ax1 = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Tax Dependency Ratio (Structural Extraction Vector)
# Proving Tokyo's revenue is decoupled from physical production
colors1 = ['#e74c3c', '#95a5a6', '#3498db'] # Red for Tokyo (Anomaly), Blue for Aichi (Normal)
ax1[0].bar(entities, tax_dependency, color=colors1, alpha=0.8, width=0.6)
ax1[0].set_title('Fig 1. Tax Dependency Ratio ($R_{tax}$)\n(Proving Structural Wealth Extraction)', fontweight='bold')
ax1[0].set_ylabel('Local Tax Dependency (%)')
ax1[0].set_ylim(0, 100)
# Add annotations
for i, v in enumerate(tax_dependency):
    ax1[0].text(i, v + 2, f"{v}%", ha='center', fontweight='bold')

# Plot 2: Fund Rotation Ratio (Metabolic Stagnation)
# Proving Tokyo as a Fiscal Capacitor
colors2 = ['#c0392b', '#7f8c8d', '#2980b9']
ax1[1].bar(entities, fund_rotation, color=colors2, alpha=0.8, width=0.6)
ax1[1].set_title('Fig 2. Fund Rotation Ratio ($R_{fund}$)\n(The Fiscal Capacitor Diagnosis)', fontweight='bold')
ax1[1].set_ylabel('Ratio (Fund Balance / Annual Revenue)')
ax1[1].set_ylim(0, 0.6)
# Add annotations
for i, v in enumerate(fund_rotation):
    ax1[1].text(i, v + 0.02, f"{v}", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

# Data Summary for Citation
print("--- Empirical Data for Chapter 7 ---")
print(f"Tokyo Local Tax Revenue: 6.69 Trillion JPY (R_tax: 74.8%)")
print(f"Tokyo Fund Balance: 4.21 Trillion JPY (Half a year of stagnant stock)")
print(f"Aichi R_tax (Control Group): 51.8% (Industrial base standard)")
```
